```python
import pandas as pd
import numpy as np

```

```python
stavbe2018 = pd.read_csv(r"clean_GURS\new_prodaja2018.csv", sep=',', encoding='utf-8')
stavbe2018.head(10)
```

```python
stavbe2018.info()
```

```python
#preparing the year column for Power BI
stavbe2018['LETO'] = pd.to_datetime("2018-01-01")
```

```python

# creating a boolean column 'NOVOGRADNJA_1' based on 'NOVOGRADNJA'
stavbe2018['novogradnja_1'] = stavbe2018['novogradnja'] == 1

# Preview the new column
stavbe2018[['novogradnja', 'novogradnja_1']].head(10)
```

```python
#Printing the number of rows with missing values in 'm2_neto' column
stavbe2018[stavbe2018['m2_neto'].isna()]
```

```python
# filling missing values in 'm2_neto' 
stavbe2018['m2_neto'].fillna(stavbe2018['m2_bruto'], inplace=True)
```

```python
def oceni_sobnost(m2):
    if pd.isna(m2):
        return None
    if m2 < 35:
        return 1
    elif m2 < 50:
        return 2
    elif m2 < 70:
        return 3
    elif m2 < 90:
        return 4
    elif m2 < 110:
        return 5
    else:
        return 6
```

```python
stavbe2018['sobnost_ocenjena'] = stavbe2018['m2_neto'].apply(oceni_sobnost)
```

```python
stavbe2018.info()
```

```python
# Droping the original 'NOVOGRADNJA' column, as we now have 'NOVOGRADNJA_1' which has the same information in a boolean format.
stavbe2018.drop(columns=['novogradnja'], inplace=True)
```

```python
#Printing the number of rows with missing values in 'm2_neto' column
stavbe2018[stavbe2018['m2_neto'].isna()]
```

```python
stavbe2018.info()
```

```python
# Uniforming the column names to lowercase
stavbe2018.columns = [col.lower() for col in stavbe2018.columns]
```

```python
#uniforming the all values in all the columns to lowercase
for col in stavbe2018.columns:
    if stavbe2018[col].dtype == 'object':
        stavbe2018[col] = stavbe2018[col].str.lower()
```

```python
stavbe2018['ime_ko'].unique()
```

```python
# checking all the empty values in the 'naselje' column
stavbe2018[stavbe2018['naselje'].isna()]
```

```python
# changing all the empty values in the 'naselje' column to 'ljubljana'
stavbe2018['naselje'].fillna('ljubljana', inplace=True)
```

```python
stavbe2018[stavbe2018['cena'].isna()]
```

```python
# droping missing values in 'cena' column row 258
stavbe2018.drop(index=258, inplace=True)
```
