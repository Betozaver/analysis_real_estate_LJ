```python
import pandas as pd
import numpy as np

```

```python
stavbe2022 = pd.read_csv(r"clean_GURS\new_prodaja2022.csv", sep=',', encoding='utf-8')
stavbe2022.head(10)
```

```python
stavbe2022.info()
```

```python
#preparing the year column for Power BI
stavbe2022['LETO'] = pd.to_datetime("2022-01-01")
```

```python

# creating a boolean column 'NOVOGRADNJA_1' based on 'NOVOGRADNJA'
stavbe2022['novogradnja_1'] = stavbe2022['novogradnja'] == 1

# Preview the new column
stavbe2022[['novogradnja', 'novogradnja_1']].sample(10)
```

```python
def oceni_sobnost(m2):
    if pd.isna(m2):
        return None
    if m2 < 35:
        return 1
    elif m2 < 50:
        return 2
    elif m2 < 70:
        return 3
    elif m2 < 90:
        return 4
    elif m2 < 110:
        return 5
    else:
        return 6
```

```python
stavbe2022['sobnost_ocenjena'] = stavbe2022['m2_neto'].apply(oceni_sobnost)
```

```python
stavbe2022.info()
```

```python
# Droping the original 'NOVOGRADNJA' column, as we now have 'NOVOGRADNJA_1' which has the same information in a boolean format.
stavbe2022.drop(columns=['novogradnja'], inplace=True)
```

```python
# Uniforming the column names to lowercase
stavbe2022.columns = [col.lower() for col in stavbe2022.columns]
```

```python
#uniforming the all values in all the columns to lowercase
for col in stavbe2022.columns:
    if stavbe2022[col].dtype == 'object':
        stavbe2022[col] = stavbe2022[col].str.lower()
```

```python
stavbe2022['ime_ko'].unique()
```

```python
# checking all the empty values in the 'naselje' column
stavbe2022[stavbe2022['naselje'].isna()]
```

```python
# changing all the empty values in the 'naselje' column to 'ljubljana'
stavbe2022['naselje'].fillna('ljubljana', inplace=True)
```

```python
stavbe2022.info()
```

```python
stavbe2022.to_csv(r"clean_GURS\stavbe2022_cleaned.csv", index=False, encoding='utf-8')
```
