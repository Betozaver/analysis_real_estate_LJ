```python
import pandas as pd
import numpy as np

```

```python
stavbe2015 = pd.read_csv(r"clean_GURS\new_prodaja2015.csv", sep=',', encoding='utf-8')
stavbe2015.head(10)
```

```python
stavbe2015.info()
```

```python
#preparing the year column for Power BI
stavbe2015['LETO'] = pd.to_datetime("2015-01-01")
```

```python

# creating a boolean column 'NOVOGRADNJA_1' based on 'NOVOGRADNJA'
stavbe2015['novogradnja_1'] = stavbe2015['novogradnja'] == 1

# Preview the new column
stavbe2015[['novogradnja', 'novogradnja_1']].head(10)
```

```python
def oceni_sobnost(m2):
    if pd.isna(m2):
        return None
    if m2 < 35:
        return 1
    elif m2 < 50:
        return 2
    elif m2 < 70:
        return 3
    elif m2 < 90:
        return 4
    elif m2 < 110:
        return 5
    else:
        return 6
```

```python
stavbe2015['sobnost_ocenjena'] = stavbe2015['m2_neto'].apply(oceni_sobnost)
```

```python
stavbe2015.info()
```

```python
# Droping the original 'NOVOGRADNJA' column, as we now have 'NOVOGRADNJA_1' which has the same information in a boolean format.
stavbe2015.drop(columns=['novogradnja'], inplace=True)
```

```python
# Droping 'NASELJE' column as it is not needed for further analysis, and also has 50 missing values.
# This column is not used in the analysis, and it has a significant number of missing values
#stavbe2014.drop(columns=['NASELJE'], inplace=True)
```

```python
#Printing the number of rows with missing values in 'm2_neto' column
stavbe2015[stavbe2015['m2_neto'].isna()]
```

```python
# filling missing values in 'm2_neto' 
stavbe2015['m2_neto'].fillna(stavbe2015['m2_bruto'], inplace=True)
```

```python
stavbe2015.info()
```

```python
# Uniforming the column names to lowercase
stavbe2015.columns = [col.lower() for col in stavbe2015.columns]
```

```python
#uniforming the all values in all the columns to lowercase
for col in stavbe2015.columns:
    if stavbe2015[col].dtype == 'object':
        stavbe2015[col] = stavbe2015[col].str.lower()
```

```python
stavbe2015['ime_ko'].unique()
```

```python
# checking all the empty values in the 'naselje' column
stavbe2015[stavbe2015['naselje'].isna()]
```

```python
# changing all the empty values in the 'naselje' column to 'ljubljana'
stavbe2015['naselje'].fillna('ljubljana', inplace=True)
```

```python
stavbe2015.info()
```
