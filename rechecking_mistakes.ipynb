```python
df = pd.read_csv(r"power_bi_ready\stavbe2024_cleaned.csv", sep=",", encoding="utf-8", 
                   parse_dates=["leto"])
df.sample(10)
```

```python
# 2. Recreate your Year and price_m2
df["leto"] = df["leto"].dt.year
df["price_m2"] = df["cena"] / df["m2_bruto"]
```

```python
# 3. Catch infinite or zero‐division cases
df["price_m2"].replace([np.inf, -np.inf], np.nan, inplace=True)
df.replace({"price_m2": {0: np.nan}}, inplace=True)
```

```python
# 4. Quick summary per year
summary = (
    df
    .groupby("leto")["m2_bruto"]
    .agg(count="count", non_null=lambda x: x.notna().sum(),
         mean="mean", min="min", max="max", median="median")
    .reset_index()
)
print(summary)
```
