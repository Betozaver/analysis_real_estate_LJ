```python
import pandas as pd
import numpy as np

```

```python
stavbe2023 = pd.read_csv(r"clean_GURS\new_prodaja2023.csv", sep=',', encoding='utf-8')
stavbe2023.sample(10)
```

```python
stavbe2023.info()
```

```python
#preparing the year column for Power BI
stavbe2023['LETO'] = pd.to_datetime("2023-01-01")
```

```python

# creating a boolean column 'NOVOGRADNJA_1' based on 'NOVOGRADNJA'
stavbe2023['novogradnja_1'] = stavbe2023['novogradnja'] == 1

# Preview the new column
stavbe2023[['novogradnja', 'novogradnja_1']].sample(10)
```

```python
#Printing the number of rows with missing values in 'm2_neto' column
stavbe2023[stavbe2023['m2_neto'].isna()]
```

```python
stavbe2023['m2_neto'].fillna(stavbe2023['m2_bruto'], inplace=True)
```

```python
def oceni_sobnost(m2):
    if pd.isna(m2):
        return None
    if m2 < 35:
        return 1
    elif m2 < 50:
        return 2
    elif m2 < 70:
        return 3
    elif m2 < 90:
        return 4
    elif m2 < 110:
        return 5
    else:
        return 6
```

```python
stavbe2023['sobnost_ocenjena'] = stavbe2023['m2_neto'].apply(oceni_sobnost)
```

```python
stavbe2023.info()
```

```python
# Droping the original 'NOVOGRADNJA' column, as we now have 'NOVOGRADNJA_1' which has the same information in a boolean format.
stavbe2023.drop(columns=['novogradnja'], inplace=True)
```

```python
stavbe2023.info()
```

```python
# Uniforming the column names to lowercase
stavbe2023.columns = [col.lower() for col in stavbe2023.columns]
```

```python
#uniforming the all values in all the columns to lowercase
for col in stavbe2023.columns:
    if stavbe2023[col].dtype == 'object':
        stavbe2023[col] = stavbe2023[col].str.lower()
```

```python
stavbe2023['ime_ko'].unique()
```

```python
# checking all the empty values in the 'naselje' column
stavbe2023[stavbe2023['naselje'].isna()]
```

```python
# changing all the empty values in the 'naselje' column to 'ljubljana'
stavbe2023['naselje'].fillna('ljubljana', inplace=True)
```

```python
stavbe2023.info()
```

```python
stavbe2023.to_csv(r"clean_GURS\stavbe2023_cleaned.csv", index=False, encoding='utf-8')
```
