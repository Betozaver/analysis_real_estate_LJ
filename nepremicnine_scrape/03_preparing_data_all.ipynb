```python
df1 = pd.read_csv(r"houses_cleaned.csv", sep=",", encoding="utf-8")
```

```python
# 1) Uniform column names
df1.columns = df1.columns.str.lower()

# 2) Identify all object-dtype columns
str_cols = df1.select_dtypes(include="object").columns

# 3) Lowercase every value in those columns
df1[str_cols] = df1[str_cols].apply(lambda col: col.str.lower())
```

```python
def oceni_sobnost(m2):
    if pd.isna(m2):
        return None
    if m2 < 35:
        return 1
    elif m2 < 50:
        return 2
    elif m2 < 70:
        return 3
    elif m2 < 90:
        return 4
    elif m2 < 110:
        return 5
    else:
        return 6
    
df1['sobnost_ocenjena'] = df1['m2'].apply(oceni_sobnost)
```

```python
df1.drop(columns=['naslov'], inplace=True)
df1.drop(columns=['opis'], inplace=True)
df1['leto_izgradnje'] = df1['leto']
df1['total_m2'] = df1['m2']
df1.drop(columns=['nastropje'], inplace=True)
df1.drop(columns=['zemljisce'], inplace=True)

df1['per_per_m2'] = df1['cena_na_m2']
df1['sobnost_ocenjena'] = df1['st_sob']
```
