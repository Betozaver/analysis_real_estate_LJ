##Loading all of the tables and joining them in one


```python
df = pd.read_csv(r"power_bi_ready\gurs_cut\realestate_2014.csv", sep=",", encoding="utf-8",)
df1 = pd.read_csv(r"power_bi_ready\gurs_cut\realestate_2015.csv", sep=",", encoding="utf-8",)
df2 = pd.read_csv(r"power_bi_ready\gurs_cut\realestate_2016.csv", sep=",", encoding="utf-8",)
df3 = pd.read_csv(r"power_bi_ready\gurs_cut\realestate_2017.csv", sep=",", encoding="utf-8",)
df4 = pd.read_csv(r"power_bi_ready\gurs_cut\realestate_2018.csv", sep=",", encoding="utf-8",)
df5 = pd.read_csv(r"power_bi_ready\gurs_cut\realestate_2019.csv", sep=",", encoding="utf-8",)
df6 = pd.read_csv(r"power_bi_ready\gurs_cut\realestate_2020.csv", sep=",", encoding="utf-8",)
df7 = pd.read_csv(r"power_bi_ready\gurs_cut\realestate_2022.csv", sep=",", encoding="utf-8",)
df8 = pd.read_csv(r"power_bi_ready\gurs_cut\realestate_2023.csv", sep=",", encoding="utf-8",)
df9 = pd.read_csv(r"power_bi_ready\gurs_cut\realestate_2024.csv", sep=",", encoding="utf-8",)
df0 = pd.read_csv(r"power_bi_ready\gurs_cut\realestate_2021.csv", sep=",", encoding="utf-8",)
```

```python
df1.sample(10)
```

```python
# merge

# 1) Put them in a list
dfs = [df, df1, df2, df3, df4, df5, df6, df7, df8, df9, df0]

# 2) Concatenate into one DataFrame
full_df = pd.concat(dfs, ignore_index=True)

# 3) (Optional) verify
print(full_df.shape)         # rows should equal sum of all individual dfs
print(full_df['leto'].unique())  # you should see [2014, …, 2024]
```

```python
# prepareing area columns and price column for Power BI
full_df['total_m2'] = full_df['total_m2'].astype(str).str.replace('.', ',', regex=False)
full_df['living_space'] = full_df['living_space'].astype(str).str.replace('.', ',', regex=False)
full_df['cena'] = full_df['cena'].astype(str).str.replace('.', ',', regex=False)
full_df['price_per_m2'] = full_df['price_per_m2'].astype(str).str.replace('.', ',', regex=False)
```

```python
full_df.sample(10)
```

```python
full_df.to_csv(r"power_bi_ready\gurs_cut\finished_full_realestate.csv", sep=",", encoding="utf-8", index=False)
```
