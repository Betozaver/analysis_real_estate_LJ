```python
import pandas as pd
import numpy as np

```

```python
stavbe2014 = pd.read_csv(r"clean_GURS\new_prodaja2014.csv", sep=',', encoding='utf-8')
stavbe2014.sample(10)
```

```python
#preparing the year column for Power BI
stavbe2014['LETO'] = pd.to_datetime("2014-01-01")
```

```python

# creating a boolean column 'NOVOGRADNJA_1' based on 'NOVOGRADNJA'
stavbe2014['novogradnja_1'] = stavbe2014['novogradnja'] == 1

# Preview the new column
stavbe2014[['novogradnja', 'novogradnja_1']].sample(10)
```

```python
#Printing the number of rows with missing values in 'm2_neto' column
stavbe2014[stavbe2014['m2_neto'].isna()]
```

```python

# filling missing values in 'm2_neto', replacing 0 with NaN
stavbe2014.loc[stavbe2014['m2_neto'] == 0, 'm2_neto'] = np.nan
```

```python


stavbe2014[stavbe2014['m2_neto'] == 0].count()
```

```python

# filling missing values in 'm2_neto' using DataFrame syntax
stavbe2014.fillna({'m2_neto': stavbe2014['m2_bruto']}, inplace=True)
```

```python
def oceni_sobnost(m2):
    if pd.isna(m2):
        return None
    if m2 < 35:
        return 1
    elif m2 < 50:
        return 2
    elif m2 < 70:
        return 3
    elif m2 < 90:
        return 4
    elif m2 < 110:
        return 5
    else:
        return 6
```

```python
stavbe2014['sobnost_ocenjena'] = stavbe2014['m2_neto'].apply(oceni_sobnost)
```

```python
stavbe2014.info()
```

```python
# Droping the original 'NOVOGRADNJA' column, as we now have 'NOVOGRADNJA_1' which has the same information in a boolean format.
stavbe2014.drop(columns=['novogradnja'], inplace=True)
```

```python
stavbe2014.info()
```

```python
# Uniforming the column names to lowercase
stavbe2014.columns = [col.lower() for col in stavbe2014.columns]
```

```python
#uniforming the all values in all the columns to lowercase
for col in stavbe2014.columns:
    if stavbe2014[col].dtype == 'object':
        stavbe2014[col] = stavbe2014[col].str.lower()
```

```python
# changing all the empty values in the 'naselje' column to 'ljubljana'
stavbe2014.fillna({'naselje': 'ljubljana'}, inplace=True)
```
