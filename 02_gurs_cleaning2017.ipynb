```python
import pandas as pd
import numpy as np

```

```python
stavbe2017 = pd.read_csv(r"clean_GURS\new_prodaja2017.csv", sep=',', encoding='utf-8')
stavbe2017.head(10)
```

```python
stavbe2017.info()
```

```python
#preparing the year column for Power BI
stavbe2017['LETO'] = pd.to_datetime("2017-01-01")
```

```python

# creating a boolean column 'NOVOGRADNJA_1' based on 'NOVOGRADNJA'
stavbe2017['novogradnja_1'] = stavbe2017['novogradnja'] == 1

# Preview the new column
stavbe2017[['novogradnja', 'novogradnja_1']].head(10)
```

```python
#Printing the number of rows with missing values in 'm2_neto' column
stavbe2017[stavbe2017['m2_neto'].isna()]
```

```python
# filling missing values in 'm2_neto' 
stavbe2017['m2_neto'].fillna(stavbe2017['m2_bruto'], inplace=True)
```

```python
def oceni_sobnost(m2):
    if pd.isna(m2):
        return None
    if m2 < 35:
        return 1
    elif m2 < 50:
        return 2
    elif m2 < 70:
        return 3
    elif m2 < 90:
        return 4
    elif m2 < 110:
        return 5
    else:
        return 6
```

```python
stavbe2017['sobnost_ocenjena'] = stavbe2017['m2_neto'].apply(oceni_sobnost)
```

```python
stavbe2017.info()
```

```python
# Droping the original 'NOVOGRADNJA' column, as we now have 'NOVOGRADNJA_1' which has the same information in a boolean format.
stavbe2017.drop(columns=['novogradnja'], inplace=True)
```

```python
#Printing the number of rows with missing values in 'm2_neto' column
stavbe2017[stavbe2017['m2_neto'].isna()]
```

```python
stavbe2017.info()
```

```python
# Uniforming the column names to lowercase
stavbe2017.columns = [col.lower() for col in stavbe2017.columns]
```

```python
#uniforming the all values in all the columns to lowercase
for col in stavbe2017.columns:
    if stavbe2017[col].dtype == 'object':
        stavbe2017[col] = stavbe2017[col].str.lower()
```

```python
stavbe2017['ime_ko'].unique()
```

```python
# checking all the empty values in the 'naselje' column
stavbe2017[stavbe2017['naselje'].isna()]
```

```python
# changing all the empty values in the 'naselje' column to 'ljubljana'
stavbe2017['naselje'].fillna('ljubljana', inplace=True)
```
