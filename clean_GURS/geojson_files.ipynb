##Loading all of the tables and joining them in one


```python
df = pd.read_csv(r"prodaja2014_202507241023.csv", sep=",", encoding="utf-8",)
df1 = pd.read_csv(r"prodaja2015_202507231453.csv", sep=",", encoding="utf-8",)
df2 = pd.read_csv(r"prodaja2016_202507231442.csv", sep=",", encoding="utf-8",)
df3 = pd.read_csv(r"prodaja2017_202507241047.csv", sep=",", encoding="utf-8",)
df4 = pd.read_csv(r"prodaja2018_202507241138.csv", sep=",", encoding="utf-8",)
df5 = pd.read_csv(r"prodaja2019_202507241149.csv", sep=",", encoding="utf-8",)
df6 = pd.read_csv(r"prodaja2020_202507241157.csv", sep=",", encoding="utf-8",)
df7 = pd.read_csv(r"prodaja2021_202507241207.csv", sep=",", encoding="utf-8",)
df8 = pd.read_csv(r"prodaja2022_202507241213.csv", sep=",", encoding="utf-8",)
df9 = pd.read_csv(r"prodaja2023_202507241310.csv", sep=",", encoding="utf-8",)
df0 = pd.read_csv(r"prodaja2024_202507241308.csv", sep=",", encoding="utf-8",)
```

```python
df1.sample(10)
```

```python
# merge

# 1) Put them in a list
dfs = [df, df1, df2, df3, df4, df5, df6, df7, df8, df9, df0]

# 2) Concatenate into one DataFrame
full_df = pd.concat(dfs, ignore_index=True)

# 3) (Optional) verify
print(full_df.shape)         # rows should equal sum of all individual dfs
print(full_df['LETO'].unique())  # you should see [2014, …, 2024]
```
